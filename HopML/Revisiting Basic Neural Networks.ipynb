# Revisiting Basic Neural Network

A feedforward ReLU neural network $\mathcal{N}$ with input $n$, output width $m$ and depth $k$ consists of 

* A sequence of integer <i>widths</i> $n=d_1, d_2, \dots, d_{k+1} = m$,

* A sequence of affine functions $\{ A_i : \mathbb{R}^{d_i} \xrightarrow{} \mathbb{R}^{d_{i+1}}\}_{i=1}^k$,

Associated to every ReLU net $\mathcal{N}$ is a continuous function

$$f_{\mathcal{N}} : \mathbb{R}^n \xrightarrow{} \mathbb{R}^m , $$

$$f_{\mathcal{N}} = A_k \circ ReLU \circ \dots \circ ReLU \circ A_1$$

where the <u>rectified linear unit</u> (ReLU) is the function

$$\hbox{ReLU}: \mathbb{R}^d \xrightarrow{} \mathbb{R}^d, $$

$$(x_1, \dots, x_d) \xrightarrow{} ( \max \{ x_1, 0\}, \dots, \max \{ x_d, 0 \})$$

## NN is a function approximator

**Theorem** (Stone-Weierstrass). If $K \subseteq \mathbb{R}^n$ is <i>compact</i>, $f : K \xrightarrow{} \mathbb{R}^m$ is <i>continuous</i> and $\epsilon > 0$, there exists a polynomial function $g : \mathbb{R}^n \xrightarrow{} \mathbb{R}^m$ such that

$$\sup_{x \in K} || f(x) - g(x) || < \epsilon$$

**Theorem** (Hanin-Sellke '17). If $K \subseteq \mathbb{R}^n$ is compact, $f : K \xrightarrow{} \mathbb{R}^m$ is continuous and $\epsilon > 0$, there exists a ReLU net $\mathcal{N}$ with widths bounded above by $m + n$ such that

$$\sup_{x \in K} ||f(x) - f_{\mathcal{N}}(x)|| < \epsilon$$

This is achievable with depth $O(diam(K) / w_f^{-1}(\epsilon))^{n+1}$.

$$w_f^{-1}(\epsilon) = \sup \{ \delta > 0 \hbox{ } | \hbox{  } ||x-y|| \leq \delta \Rightarrow ||f(x) - f(y)|| \leq \epsilon \}$$

## Examples

ReLU net $\mathcal{N} = \bigl\{ A_i : \mathbb{R}^{d_i} \xrightarrow{} \mathbb{R}^{d_{i+1}} \bigr\}_{i=1}^k \in \Pi_{i=1}^k M_{d_{i+1}, d_i} (\mathbb{R}) \times \Pi_{i=1}^k \mathbb{R}^{d_{i+1}}$

$$f_{\mathcal{N}} : \mathbb{R}^n \xrightarrow{} \mathbb{R}^m,$$

$$f_{\mathcal{N}} = A_k \circ ReLU \circ \dots \circ ReLU \circ A_1$$
